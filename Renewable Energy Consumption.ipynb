### Cleaning Data

#### Some entry in the country name column are not names of countries but are in fact region names or groups of countries.

#### Code
Drop entries in the country name column that are not names of countries.

```python
df_clean = df.copy()
```

```python
drop_rows = ['Africa Eastern and Southern', 'Arab World', 'Africa Western and Central', 'Central Europe and the Baltics', 
             'Caribbean small states', 'East Asia & Pacific (excluding high income)', 
             'Early-demographic dividend', 'East Asia & Pacific', 'Europe & Central Asia (excluding high income)', 
             'Europe & Central Asia', 'Euro area', 'European Union', 'Fragile and conflict affected situations', 
             'High income', 'Heavily indebted poor countries (HIPC)', 'IBRD only', 'IDA & IBRD total', 'IDA total', 
             'IDA blend', 'IDA only', 'Not classified', 'Latin America & Caribbean (excluding high income)', 
             'Latin America & Caribbean', 'Least developed countries: UN classification', 'Low income', 
             'Lower middle income', 'Low & middle income', 'Late-demographic dividend', 'Middle East & North Africa', 
             'Middle income', 'Middle East & North Africa (excluding high income)', 'North America', 'OECD members', 
             'Other small states', 'Pre-demographic dividend', 'Pacific island small states', 'Post-demographic dividend', 
             'Sub-Saharan Africa (excluding high income)', 'Sub-Saharan Africa', 'Small states', 
             'East Asia & Pacific (IDA & IBRD countries)', 'Europe & Central Asia (IDA & IBRD countries)', 
             'Latin America & the Caribbean (IDA & IBRD countries)', 'Middle East & North Africa (IDA & IBRD countries)', 
             'South Asia (IDA & IBRD)', 'Sub-Saharan Africa (IDA & IBRD countries)', 'Upper middle income', 'World', 
             'South Asia']
```

```python
non_country_names = df_clean[df_clean['Country Name'].isin(drop_rows)]
non_country_names
```

```python
idx = list(non_country_names.index)
```

```python
df_clean = df_clean.drop(idx)
```

#### Wrong spelling of some country names = ['Yemen, Rep.', 'Venezuela, RB', 'Korea, Rep.', 'Iran, Islamic Rep.', 'Gambia, The', 'Egypt, Arab Rep.', 'Congo, Rep.', 'Congo, Dem. Rep.', 'Bahamas, The', ]

#### Code
Replace wrongly spelt country names with the right spelling.

```python
incorrect_country_names = ['Yemen, Rep.', 'Venezuela, RB', 'Korea, Rep.', 'Iran, Islamic Rep.', 'Gambia, The', 
                           'Egypt, Arab Rep.', 'Congo, Rep.', 'Congo, Dem. Rep.', 'Bahamas, The']
```

```python
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Yemen, Rep.', 'Yemen Rep.')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Venezuela, RB', 'Venezuela RB')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Korea, Rep.', 'Korea Rep.')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Iran, Islamic Rep.', 'Iran Islamic Rep.')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Gambia, The', 'Gambia')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Egypt, Arab Rep.', 'Egypt Arab Rep.')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Congo, Rep.', 'Congo Rep.')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Congo, Dem. Rep.', 'Congo Dem. Rep.')
df_clean['Country Name'] = df_clean['Country Name'].str.replace('Bahamas, The', 'The Bahamas')
```

#### The year columns (1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018) should infact be one column; the Energy consumption column.

#### Code
Use the pd.melt() function to convert the year columns to one column containing the energy consumption values.

```python
df_clean = pd.melt(df_clean, id_vars = ['Country Name', 'Country Code', 'Region', 'IncomeGroup', 'Indicator Name', 'Indicator Code'], 
                  var_name = 'Year', value_name = 'Energy_Consump.')
```
